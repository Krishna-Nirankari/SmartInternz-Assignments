2. Write a function to Calculate Area and Perimeter of Rectangle

```python
def area(length, breadth):
  area = length * breadth
  return area

def perimeter(length, breadth):
  perimeter = 2 * (length + breadth)
  return perimeter

length = 5
breadth = 10

print(f"Area: {area(length,breadth)}")
print(f"Perimeter: {perimeter(length,breadth)}")
```

5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)

```python
def calculate_electricity_bill(units):
    total_bill = 0

    # For the first 100 units
    if units <= 100:
        total_bill += 0
    else:
        total_bill += (units - 100) * 0  # No charge for the first 100 units

    # For the next 100 units
    if units > 100 and units <= 200:
        total_bill += (units - 100) * 5
    elif units > 200:
        total_bill += 100 * 5

    # For units above 200
    if units > 200:
        total_bill += (units - 200) * 10

    return total_bill

# Accept the number of units from the user
units = int(input("Enter the number of units: "))

# Calculate the electricity bill
bill_amount = calculate_electricity_bill(units)

# Display the bill amount
print("Total bill amount:", bill_amount)
```

7. Write a python program to check given word is Palindrome or not

```python
word = input("Enter a word: ")
word = word.lower()
reversed_word = word[::-1]

if word == reversed_word:
  print("The word is a palindrome.")
else:
  print("The word is not a palindrome.")
```

9. Perform Univariate Analysis

```python
import matplotlib.pyplot as plt
# Histogram for 'Area'
plt.hist(df['lot area'], bins=10)
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.title('Distribution of Area')
plt.show()
```

```python
# Bar chart for 'Availability'
df['number of bedrooms'].value_counts().plot(kind='bar')
plt.xlabel('number of bedrooms')
plt.ylabel('Count')
plt.title('number of bedrooms')
plt.show()
```

```python
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.hist(df['number of bedrooms'], bins=20, color='magenta', edgecolor='black')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Bedrooms')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.bar(df['condition of the house'].value_counts().index, df['condition of the house'].value_counts().values, color='salmon')
plt.xlabel('Condition of the House')
plt.ylabel('Count')
plt.title('Bar Chart of Condition of the House')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```
